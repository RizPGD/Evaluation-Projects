As per the above observation,  for column "Number_Weeks_Used" there are nearly 10% of data is of Null value which will not impact our train dataset. We will drop the rows which are null such that the whole dataset will not be impacted.

```python
#Removing the Null value from the train dataset.

df_train1=df_train.dropna(subset=['Number_Weeks_Used'])
df_train1.head()
```

```python
sns.heatmap(df_train1.isnull())
```

All the Null Values are removed.

We can remove the "ID" column from our training dataset as it will not be useful for the Prediction model.

```python
df_train1=df_train1.drop(['ID'],axis=1,errors='ignore')
df_train1.head()
```

```python
df_train1.corr()
```

```python
# Let us check the Skewness in the dataset
df_train1.skew()
```

```python
sns.distplot(df_train1.skew())
```

```python
#There are some skewness in our train dataset. We will try to normalize the data.

#Applying Log Transform Method
for col in df_train1.columns:
    if df_train1.skew().loc[col]>0.55:
        df_train1[col]=np.log1p(df_train1[col])
df_train1.skew()
```

```python
#We will try Square Root Transform method 
df_train1_sqrt=np.sqrt(df_train1)
df_train1_sqrt.skew()
```

```python
#Now Let us check for the Outliers in our dataset

# Applying IQR method to find the outliers

Q1=df_train1_sqrt.quantile(0.25)
Q3=df_train1_sqrt.quantile(0.75)

IQR=Q3-Q1

print(IQR)
```

```python
print(df_train1_sqrt < (Q1-1.5*IQR)) 
print(df_train1_sqrt > (Q3+1.5*IQR))
```

```python
df_train2=df_train1_sqrt[~((df_train1_sqrt<(Q1-1.5*IQR)) | (df_train1_sqrt>(Q3+1.5*IQR)))]
print(df_train2.shape)
print(df_train1.shape)
```

There is no outlier in our dataset.So we can go for the visualization of the data

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df_train1.Estimated_Insects_Count=le.fit_transform(df_train1.Estimated_Insects_Count)
df_train1.Crop_Type=le.fit_transform(df_train1.Crop_Type)
df_train1.Soil_Type=le.fit_transform(df_train1.Soil_Type)
df_train1.Pesticide_Use_Category=le.fit_transform(df_train1.Pesticide_Use_Category)
df_train1.Number_Doses_Week=le.fit_transform(df_train1.Number_Doses_Week)
df_train1.Number_Weeks_Used=le.fit_transform(df_train1.Number_Weeks_Used)
df_train1.Number_Weeks_Quit=le.fit_transform(df_train1.Number_Weeks_Quit)
df_train1.Season=le.fit_transform(df_train1.Season)
df_train1.Crop_Damage=le.fit_transform(df_train1.Crop_Damage)

```

```python

X=df_train1.iloc[:,:-1]
Y=df_train1.iloc[:,-1]
print(X)
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler

Standsc=StandardScaler()
X=Standsc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=20)
```

```python
#Applying RandomForest Classifuier model for prediction

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

agri_model=RandomForestClassifier()
agri_model.fit(X_train,Y_train)
```

```python
#Prediction for test model

Y_pred= agri_model.predict(X_test)
```

```python
#Accuracy Testing:

accuracy_agri=accuracy_score(Y_test,Y_pred)
accuracy_agri
```

Accuracy score using Random forest Classifier model is 82%

```python
# We will apply LogisticRegression to our model to check for the Model accuracy

from sklearn.linear_model import LogisticRegression

logreg_agri=LogisticRegression()
logreg_agri.fit(X_train,Y_train)

pred= logreg_agri.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

accuracy_agri=accuracy_score(Y_test,pred)
accuracy_agri
```

Accuracy Score by using LogisticRegression is 83%
