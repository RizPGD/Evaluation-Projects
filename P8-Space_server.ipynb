# Predictive Modeling

```python
x=df_train2
print(x)
print(y)
```

We have separated the Input and Target variables from the given dataset.As we can check all the columns are continuous variables.

We will apply the Classification model to predict the best result for our target variables.

```python
from sklearn.preprocessing import StandardScaler

Standsc=StandardScaler()
x=Standsc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=9)
```

```python
print(x_train.shape,x_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

space_model=RandomForestClassifier()
space_model.fit(x_train,y_train)
```

```python
y_pred= space_model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.model_selection import cross_val_score

Cross_val =cross_val_score(space_model,x,y,cv=10,scoring='accuracy').mean()
Cross_val
```

We are getting the Accuracy score of 1  for our predictive model with RandomForestClassifier method and we have cross varified with cross_val_score

Our model is working fine with for the predictiuon of Space object category.
