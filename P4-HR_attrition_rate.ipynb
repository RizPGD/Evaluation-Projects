```python
# We need to remove the categorlical values colomn to remove the skewness from the continuous dataset

df_ibm1=df_ibm.drop(['Attrition','BusinessTravel','Department','EducationField','Gender','MaritalStatus','JobRole','Over18','OverTime',''],axis=1,errors='ignore')
df_ibm1
```

Now, we will check for the outliers in our dataset.

```python
# Applying Zscore method to predict the outliers in dataset
from scipy import stats
import numpy as np

z_ibm=np.abs(stats.zscore(df_ibm1))
print(z_ibm)
```

```python
threshold=3
print(np.where(z_ibm>3))
```

```python
# Using the IQR method to detect the outlier in the dataset

Q1=df_ibm1.quantile(0.25)
Q3=df_ibm1.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
print(df_ibm1 < (Q1-1.5*IQR)) 
print(df_ibm1 > (Q3+1.5*IQR))
```

```python
df_ibm_out=df_ibm1[~((df_ibm1<(Q1-1.5*IQR)) | (df_ibm1>(Q3+1.5*IQR)))]
print(df_ibm1.shape)
print(df_ibm_out.shape)
```

From the above observation, it is clear that there are  no outliers present in out dataset.

We have many Categorical values in our dataset.So, we will use Label Encoder method to change the Categorical data into numerical data.

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df_ibm.Attrition=le.fit_transform(df_ibm.Attrition)
df_ibm.BusinessTravel=le.fit_transform(df_ibm.BusinessTravel)
df_ibm.EducationField=le.fit_transform(df_ibm.EducationField)
df_ibm.Gender=le.fit_transform(df_ibm.Gender)
df_ibm.Department=le.fit_transform(df_ibm.Department)
df_ibm.JobRole=le.fit_transform(df_ibm.JobRole)
df_ibm.MaritalStatus=le.fit_transform(df_ibm.MaritalStatus)
df_ibm.Over18=le.fit_transform(df_ibm.Over18)
df_ibm.OverTime=le.fit_transform(df_ibm.OverTime)

```

```python
df_ibm.head()
```

# Data Training and Testing

```python
from sklearn.model_selection import train_test_split

x=df_ibm.iloc[:,1:]
print(x)
y=df_ibm.iloc[:,1]
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

ibm_model=RandomForestClassifier()
ibm_model.fit(x_train,y_train)
```

```python
#Prediction for test model

y_pred= ibm_model.predict(x_test)
```

```python
#Accuracy Testing:

accuracy_ibm=accuracy_score(y_test,y_pred)
accuracy_ibm
```

# As per the above Accuracy score, we can conclude that the Attrition of Comapny Employees totally dependend on the features like Employess working hours,Jobe Roles, Job Involvement,Job Satisfaction,Monthly Income and variuos other features.
